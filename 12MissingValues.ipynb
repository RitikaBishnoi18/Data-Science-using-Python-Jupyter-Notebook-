```python
import numpy as np 
import pandas as pd
```

```python
df=pd.read_excel("C:/Users/Riya/Downloads/MissingData.xlsx")
df.head(15)
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
print(df["Salary"])
```

```python
df["Salary"]=df["Salary"].replace(np.nan,df["Salary"].mean())
print(df["Salary"])
```

```python
df=df.fillna(method="ffill")
df.head(15)# shows email as Nan bcoz has nothing above it to fill
```

```python
df=df.fillna(method="bfill")
```

```python
df["Salary"]=df["Salary"].round(2)
df.head(11)
```

```python
df["Last Name"]=df["Last Name"].replace("J","Surname")
df.head(11)
```

```python
df.isnull().sum()
```
