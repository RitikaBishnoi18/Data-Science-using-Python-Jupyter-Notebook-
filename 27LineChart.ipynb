```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/Riya/Downloads/AirPassengers.csv",names=["Month","passengers"],header=0,parse_dates=["Month"])
df
```

```python
df.info()
```

```python
import datetime as dt
```

```python
#df.groupby(dt.date(df.Month).year).passengers.sum()
df["Work Year"]=df["Month"].apply(lambda x: x.replace(month=1,day=1))
df
```

```python
g=df.groupby("Work Year").passengers.sum()
g
```

```python
plt.figure(figsize=(12,4))
plt.plot(g.index,g.values)
plt.title("Time Series")
plt.ylabel("Number of Passengers")
plt.xlabel("Year")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.plot(g.index,g.values,color="m",marker="o",ls="--")
plt.title("Time Series")
plt.ylabel("Number of Passengers")
plt.xlabel("Year")
plt.show()
```

```python
h=df.groupby("Work Year").passengers.mean()
plt.figure(figsize=(12,4))
plt.plot(g.index,g.values,color="m",marker="o",ls="--",label="Total Passengers")
plt.plot(h.index,h.values,color="g",marker="^",label="Average Passengers")
plt.title("Time Series")
plt.ylabel("Number of Passengers")
plt.xlabel("Year")
plt.legend(loc=0,ncol=2)#by default 0 ->find best loc itself
plt.show()
```

```python
h=df.groupby("Work Year").passengers.mean()
plt.figure(figsize=(12,4))
plt.plot(g.index,g.values,color="m",marker="o",ls="--",label="Total Passengers")
plt.plot(h.index,h.values,color="g",marker="^",label="Average Passengers")
plt.title("Time Series")
plt.ylabel("Number of Passengers")
plt.xlabel("Year")
plt.legend(loc=0,bbox_to_anchor=(0.8,1.2),ncol=2)#by default 0 ->find best loc itself
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.plot(g.index,g.values,color="m",marker="o",ls="--",label="Total Passengers")
plt.plot(h.index,h.values,color="g",marker="^",label="Average Passengers",alpha=0.5)
plt.title("Time Series")
plt.ylabel("Number of Passengers")
plt.xlabel("Year")
plt.legend(loc=0)#by default 0 ->find best loc itself
plt.show()
```
