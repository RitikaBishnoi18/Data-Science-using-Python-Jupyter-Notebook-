```python
gf=df.groupby("Species").SepalLengthCm.sum()
plt.pie(gf.values,labels=gf.index,explode=[0.1,0.05,0],shadow=True,autopct="%.2f%%",colors=["m","b","g"])
plt.title("Species wise Distribution")
plt.show()
```
