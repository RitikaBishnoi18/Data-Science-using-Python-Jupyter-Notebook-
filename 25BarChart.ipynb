```python
from matplotlib import pyplot as plt
```

```python
import pandas as pd 
```

```python
plt.bar(df.Species,df.SepalLengthCm)
```

```python
plt.bar(df.Species,df.SepalLengthCm)
plt.show()
```

```python
df.groupby("Species").SepalLengthCm.sum()
```

```python
g=df.groupby("Species")
gf=g.SepalLengthCm.sum()
#gf=df.groupby("Species").SepalLengthCm.sum()
plt.bar(gf.index,gf.values)#"index" show species group value and "values" show aggerate sum of sepallengthcm corresponding to group val of species
plt.bar_label(plt.gca().containers[0],label_type="edge")#get the current axis and starting from 0 index one by one in container of bars get labels
plt.show()
```

```python
gf=df.groupby("Species").SepalLengthCm.count()
plt.bar(gf.index,gf.values,color="orange",align="edge")
plt.bar_label(plt.gca().containers[0],label_type="center")
plt.title("Species v/s Sepal Length")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

```python
gf=df.groupby("Species").SepalLengthCm.count()
plt.figure(figsize=(12,4))#first define the figsize and then tell what kind of chart you want to build then only you can see that size
plt.barh(gf.index,gf.values,color="yellow",align="edge")
plt.title("Species v/s Sepal Length",fontsize=18,color="red")
plt.xlabel("Sepal Length (cm)",loc="right")
plt.ylabel("Species",loc="top",rotation="horizontal")
plt.show()
```

```python
gf=df.groupby("Species")["SepalLengthCm"].sum()
pf=df.groupby("Species")["PetalLengthCm"].sum()
plt.bar(gf.index,gf.values,color="orange",align="edge")
#plt.bar_label(plt.gca().containers[0],label_type="center")
plt.bar(pf.index,pf.values,color="yellow",align="edge")
#plt.bar_label(plt.gca().containers[0],label_type="center") not fine for stacked 
plt.title("Species v/s Sepal Length")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.show()
```
