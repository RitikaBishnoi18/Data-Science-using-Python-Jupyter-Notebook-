```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/Riya/Downloads/AirPassengers.csv",names=["Month","passengers"],header=0,parse_dates=["Month"])
df
```

```python
import datetime as dt
```

```python
df["Work Year"]=df["Month"].apply(lambda x: x.replace(month=1,day=1))
df
```

```python
g=df.groupby("Work Year").passengers.sum()
g
```

```python
plt.figure(figsize=(12,4))
plt.stem(g.values)
plt.title("Time Series")
plt.xlabel("Year")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.plot(g.index,g.values,color="m",marker="o",ls="--")
plt.title("Time Series")
plt.ylabel("Number of Passengers")
plt.xlabel("Year")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.stem(g.values)
plt.plot(g.values)
plt.title("Time Series")
plt.xlabel("Year")
plt.savefig("C:/Users/Riya/Downloads/stem.png",bbox_inches="tight", pad_inches=0.2,facecolor='yellow', edgecolor='pink')#always before show
plt.show()
```
