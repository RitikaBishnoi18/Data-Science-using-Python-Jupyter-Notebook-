```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/Riya/Downloads/archive (4)/adult.data.csv")
print(df)
```

```python
df.info()
```

```python
g1=df.groupby("sex")["hours-per-week"].mean()
g2=df.groupby("marital-status")["hours-per-week"].mean()
g3=df.groupby("race")["hours-per-week"].mean()
```

```python
plt.plot(g1.index,g1.values)
plt.show()
plt.plot(g2.index,g2.values)
plt.show()
plt.plot(g3.index,g3.values)
plt.show()
```

```python
plt.subplot(1,3,1)
plt.title("1st")
plt.plot(g1.index,g1.values)

plt.subplot(1,3,2)
plt.title("2nd")
plt.plot(g2.index,g2.values)

plt.subplot(1,3,3)
plt.title("3rd")
plt.plot(g3.index,g3.values)
plt.suptitle("ALL")
plt.show()
```
